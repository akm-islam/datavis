```python
def item_process(data2):
    dict1={}
    arr_columns=[]
    arr=[]
    for columns in data2.columns:
        print(columns)
        #arr_columns.append(columns)
        #init_row=''
        #print(columns)
        for rows in data2[columns]: 
            if rows not in dict1:
                dict1[rows]=1
            else:
                dict1[rows]=dict1[rows]+1
            #print(rows)
            arr.append(rows)
    for key in dict1:
        if dict1[key]>1:
            print(key,dict1[key])
```
